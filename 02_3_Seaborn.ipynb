# Seaborn
Seaborn is a statistical plotting library.

It is built on top of matplotlib but uses a simpler one line syntax.

Can directly work with python dataframes.

Easy to use, but less customization possible as compared to matplotlib.

# Types of plots:

1. Scatter plot : Relation between two continous variables (Trends, Crrelations)

2. Distribution plots (Imp) : How a single variable is distributed (patterns, skew, outliers) (Histogram, KDE plot)

3. Categorical plots: Categorical variables and their relationships with continouse data ( Box plot, bar plot, count plot)

4. Comparison plot:

5. Matrix plots: (Heatmap)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
plt.figure(figsize=(16,14))
```

```python
# 1. Scatter plot

df['gender_mod'] = df['gender'].map({0:'Female', 1:'Male'})
sns.scatterplot(x='age', y='chol', data=df, hue = 'gender_mod', alpha = 0.7)
plt.title('Age Vs Cholestrol')
plt.figure(figsize=(50,20))
```

```python
# 2. Histogram with KDE

sns.histplot(df['trestbps'], kde = True, color = 'green')
plt.title('Resting BP Distribution')
plt.figure(figsize=(50,20))
```

```python
# 3. Box plot

sns.boxplot(data = df, y='thalach', x = 'target', palette = 'Set2')
plt.title('Max Heart Rate Vs Target')
plt.figure(figsize=(50,20))
```

```python
# 4. Line plot
sns.lineplot(data = df, x = 'age', y = 'chol', hue = 'target')
plt.title('Age Vs Cholestrol Line')
plt.figure(figsize=(50,30))
```

```python
# 5. Bar Plot ( Imporved count plot for 'cp' feature)

sns.barplot(data = df, x = 'cp', y = 'chol', hue = 'target')
plt.title('Chest Pain Vs Cholestrol')
plt.figure(figsize=(50,30))
```

```python
# 6. Heatmap ( Correlation matrix)

relevant_columns = [ 'age','chol','thalach','trestbps','oldpeak','target']
correlation_matrix = df[relevant_columns].corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f', square = 'True', linewidths=0.5)
plt.title('Correlation Matrix')
plt.figure(figsize=(100,50))
#
```

```python
# 7. Pair plot ( added as a seperate figure)

sns.pairplot(df[relevant_columns], hue = 'target', palette='husl', corner = True, diag_kind = 'kde')
plt.suptitle('Pair Plot of selected features', y = 1.02)
plt.figure(figsize=(100,50))
```
