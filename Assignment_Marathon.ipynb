# Assignment

Find the correlation b/w pace and official time column of the combined marathon datasets draw scatter plot and heat map

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Combining datasets
df_2015 = pd.read_csv('marathon_results_2015.csv')
df_2016 = pd.read_csv('marathon_results_2016.csv')
df_2017 = pd.read_csv('marathon_results_2017.csv')

print(df_2015.head())
print(df_2016.head())
print(df_2017.head())
```

```python
#Concatenate the datasets into a single DataFrame
df = pd.concat([df_2015, df_2016, df_2017])
```

```python
# Convert the pace and offical time str split by ':' and
df['Pace'] = df['Pace'].str.split(':')
df['Official Time'] = df['Official Time'].str.split(':')
```

```python
# we get HH:MM:SS then convert HH into sec by multi by 3600 and MM into sec by multi by 60 and then add converted HH,MM,SS and store it
df['Pace'] = df['Pace'].apply(lambda x: int(x[0]) * 3600 + int(x[1]) * 60 + int(x[2]))
df['Official Time'] = df['Official Time'].apply(lambda x: int(x[0]) * 3600 + int(x[1]) * 60 + int(x[2]))
```

```python
# Correlation between pace and official time columns
corr = df[['Pace','Official Time']].corr()
print(corr)
```

```python
plt.scatter(df['Pace'], df['Official Time'])
plt.xlabel('Pace')
plt.ylabel('Official Time')
plt.show()
```

```python
# Heat map
sns.heatmap(corr, annot=True)
plt.show()
```
