# Matplotlib

To write code in matplotlib there are two syntax:
1. Functional
2. OOP

eg: plt.plot(x,y)  :- plotting using a simple functional call

  fig, ax = plt.subplots()  :- Create a figure and a set of subplots (Store in variable)

Main Goals:

Plot a functional relationship, eg: y = 2x

Main Types of Plots:

Line Plot : Great for showing functional relationships and continous data

Scatter plot : Useful for plotting raw data points and understanding the correlation between two variables

Bar Plot : Useful for categorical data to show comparisons between different groups

Histogram : Good for showing the distributions of a single variable

Pie Chart: Used for showing proportions or percentages of categories



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("heart.csv")
```

```python
# Create a subplot layout to accomodate different types of plots

plt.figure(figsize=(30,50))
```

```python
# 1. Scatter Plot
plt.subplot(3,2,1)
plt.scatter(df['age'], df['chol'])
plt.title("Age vs Cholesterol")
plt.xlabel("Age")
plt.ylabel("Cholesterol")
```

```python
# 2. Histogram
plt.subplot(3,2,2)
plt.hist(df['trestbps'], bins = 20)
plt.title('Resting blood pressure')
plt.xlabel('Resting blood pressure')
plt.ylabel('Frequency')
```

```python
# 3. Box plot
plt.subplot(3,2,3)
plt.boxplot(df['thalach'])
plt.title('Max heart rate')
plt.xlabel('Max heart rate')
plt.ylabel('Frequency')
```

```python
# 4. Line plot
plt.subplot(3,2,4)
df_sorted = df.sort_values(by = 'age') # sort by age for the line plot
plt.plot(df_sorted['age'], df_sorted['chol'], linestyle = '--', color = 'b')
plt.title('Age vs Cholesterol Line')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
```

```python
# 5. Bar plot
plt.subplot(2,4,8)
df['gender'].value_counts().plot(kind = 'bar')
plt.title('Gender Distribution')
```

```python
from enum import auto
# 6. Pie Plot
plt.subplot(3,2,6)
df['target'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%', startangle= 90)
plt.title('Heart Disease Distribution')

plt.tight_layout()
plt.show()
```
