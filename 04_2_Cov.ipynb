```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('weight-height.csv')
print(df.head())
```

```python
# Separate data based on gender
df_male = df[df['Gender'] == 'Male']
df_female = df[df['Gender'] == 'Female']
```

```python
# Calculate correlation
corr_overall = df[['Height', 'Weight']].corr()
corr_male = df_male[['Height', 'Weight']].corr()
corr_female = df_female[['Height', 'Weight']].corr()
```

```python
# calculate covariance
cov_overall = df[['Height', 'Weight']].cov()
cov_male = df_male[['Height', 'Weight']].cov()
cov_female = df_female[['Height', 'Weight']].cov()
```

```python
print("Overall Correlation between Height and Weight:")
print(corr_overall)
```

```python
print("Male Correlation between Height and Weight:")
print(corr_male)
```

```python
print("Female Correlation between Height and Weight:")
print(corr_female)
```

```python
print("Overall Covariance between Height and Weight:")
print(cov_overall)
```

```python
print("Male Covariance between Height and Weight:")
print(cov_male)
```

```python
print("Female Covariance between Height and Weight:")
print(cov_female)
```

# Create scatter plots with correlation values

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Overall data plot
axes[0].scatter(df['Height'], df['Weight'], color='Gold', alpha=0.5)
axes[0].set_title(f'Overall Data\nCorrelation: {corr_overall.loc["Height", "Weight"]:.2f}')
axes[0].set_xlabel('Height')
axes[0].set_ylabel('Weight')
axes[0].grid(True)

# Male data plot
axes[1].scatter(df_male['Height'], df_male['Weight'], color='Blue', alpha=0.5)
axes[1].set_title(f'Male Data\nCorrelation: {corr_male.loc["Height", "Weight"]:.2f}')
axes[1].set_xlabel('Height')
axes[1].set_ylabel('Weight')
axes[1].grid(True)

# Female data plot
axes[2].scatter(df_female['Height'], df_female['Weight'], color='pink', alpha=0.5)
axes[2].set_title(f'Female Data\nCorrelation: {corr_female.loc["Height", "Weight"]:.2f}')
axes[2].set_xlabel('Height')
axes[2].set_ylabel('Weight')
axes[2].grid(True)
```

# Create scatter plots with covariance values

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Overall data
axes[0].scatter(df['Height'], df['Weight'], color='Gold', alpha=0.5)
axes[0].set_title(f'Overall Data\nCovariance: {cov_overall.loc["Height", "Weight"]:.2f}')
axes[0].set_xlabel('Height')
axes[0].set_ylabel('Weight')
axes[0].grid(True)

# Male data
axes[1].scatter(df_male['Height'], df_male['Weight'], color='Blue', alpha=0.5)
axes[1].set_title(f'Male Data\nCovariance: {cov_male.loc["Height", "Weight"]:.2f}')
axes[1].set_xlabel('Height')
axes[1].set_ylabel('Weight')
axes[1].grid(True)

# Female data
axes[2].scatter(df_female['Height'], df_female['Weight'], color='pink', alpha=0.5)
axes[2].set_title(f'Female Data\nCovariance: {cov_female.loc["Height", "Weight"]:.2f}')
axes[2].set_xlabel('Height')
axes[2].set_ylabel('Weight')
axes[2].grid(True)
```
