```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
# Generate data with zero skew
data_zero_skew = np.random.normal(loc =0, scale = 1, size = 1000)

# Generating data with right skew
data_right_skew = np.random.exponential(scale=1, size = 1000)

# Generating data with left skew
data_left_skew = -np.random.exponential(scale=1, size = 1000)
```

```python
# mean of all these data
mean_zero_skew = np.mean(data_zero_skew)
mean_right_skew = np.mean(data_right_skew)
mean_left_skew = np.mean(data_left_skew)
```

```python
# median of all these data
median_zero_skew = np.median(data_zero_skew)
median_right_skew = np.median(data_right_skew)
median_left_skew = np.median(data_left_skew)
```

Zero-Skewed Data Plot

```python
# Plot histograms

plt.figure(figsize=(15,5))

plt.hist(data_zero_skew, bins = 30, alpha = 0.7, label = 'Zero Skew', color = 'gold')
plt.axvline(mean_zero_skew, color = 'red', linestyle = 'dashed', linewidth = 2, label = f'Mean: {mean_zero_skew:.2f}')
```

Left Skewed Data Plot

```python
plt.figure(figsize=(15,5))

plt.hist(data_left_skew, bins = 30, alpha = 0.7, label = 'left Skew', color = 'gold')
plt.axvline(mean_left_skew, color = 'red', linestyle = 'dashed', linewidth = 2, label = f'Mean: {mean_zero_skew:.2f}')
```

Right Skewed Data plot

```python
plt.figure(figsize=(15,5))

plt.hist(data_right_skew, bins = 30, alpha = 0.7, label = 'Right Skewed Data', color = 'gold')
plt.axvline(mean_right_skew, color = 'red', linestyle = 'dashed', linewidth = 2, label = f'Mean: {mean_right_skew:.2f}')
```
