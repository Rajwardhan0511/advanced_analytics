# Assignment
for the SP500 oil gold bitcoin.csv dataset:
1. Find the correlation and covariance between BITCOIN, Gold and S&P500. Print the correlation matrix and covariance matrix.
2. Draw scatter plots and heatmaps for bitcoin-gold and bitcoin-s&p500 combinations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SP500 oil gold bitcoin.csv")
df.head()
```

```python
correl = df[['BITCOIN', 'Gold', 'S&P500']].corr()
print(correl)
```

```python
covar = df[['BITCOIN', 'Gold', 'S&P500']].cov()
print(covar)
```

```python
# Scatter plot for bitcoin-gold and bitcoin-s&p500
plt.figure(figsize=(18, 5))
plt.scatter(df['BITCOIN'], df['Gold'], label='Bitcoin-Gold')
plt.scatter(df['BITCOIN'], df['S&P500'], label='Bitcoin-S&P500')
```

```python
# heatmap for bitcoin-gold and bitcoin-s&p500
sns.heatmap(correl, annot=True, cmap='coolwarm')
```
